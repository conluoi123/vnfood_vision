### 3. Khởi tạo Weights & Biases (W&B)
Đăng nhập và tạo project để theo dõi thí nghiệm.

```python
import wandb

# Đăng nhập (Sẽ yêu cầu API Key lần đầu)
wandb.login()

# Khởi tạo run
run = wandb.init(
    project="viet-food-recognition",
    entity="your-team-name", # Thay bằng tên team hoặc username của bạn
    config={
        "learning_rate": 0.001,
        "epochs": 10,
        "batch_size": 32,
        "architecture": "ResNet50"
    }
)

# Log một số thông tin test
wandb.log({"acc": 0.8, "loss": 0.2})

# Kết thúc thí nghiệm
wandb.finish()
```
