### Thống kê số lượng ảnh trong từng thư mục (Lớp)

```python
import os

# Danh sách các đuôi tệp ảnh phổ biến
image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.webp')
stats = {}

if os.path.exists(data_path):
    # Lấy danh sách thư mục gốc (raw, crawled, v.v.)
    root_subfolders = [f for f in os.listdir(data_path) if os.path.isdir(os.path.join(data_path, f))]

    for root_folder in root_subfolders:
        root_full_path = os.path.join(data_path, root_folder)
        image_count = 0

        # Sử dụng os.walk để quét đệ quy vào sâu nhất
        for root, dirs, files in os.walk(root_full_path):
            for file in files:
                if file.lower().endswith(image_extensions):
                    image_count += 1

        stats[root_folder] = image_count

    print("{:<20} | {:<10}".format("Thư mục gốc", "Tổng số ảnh (đệ quy)"))
    print("-" * 45)
    for folder, count in stats.items():
        print("{:<20} | {:<10}".format(folder, count))

    total_images = sum(stats.values())
    print("-" * 45)
    print(f"Tổng cộng tất cả các cấp: {total_images} ảnh.")
else:
    print("Không tìm thấy đường dẫn dữ liệu.")
```

### Kiểm tra cấu trúc cây và số lượng ảnh chi tiết trong thư mục `processed`

```python
import os

target_path = os.path.join(data_path, 'processed')
image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.webp')

if os.path.exists(target_path):
    print(f"Cấu trúc thư mục và số lượng ảnh trong: {target_path}\n")

    for root, dirs, files in os.walk(target_path):
        # Đếm số ảnh trong thư mục hiện tại
        images_in_folder = [f for f in files if f.lower().endswith(image_extensions)]
        count = len(images_in_folder)

        # Tính độ sâu để thụt lề (visualize tree)
        level = root.replace(target_path, '').count(os.sep)
        indent = ' ' * 4 * level

        # Tên thư mục hiện tại
        folder_name = os.path.basename(root)

        if count > 0:
            print(f"{indent}├── {folder_name}/ ({count} ảnh)")
        else:
            # Chỉ in tên thư mục nếu nó có thư mục con hoặc là thư mục gốc processed
            if dirs or root == target_path:
                print(f"{indent}└── {folder_name}/")
else:
    print(f"Thư mục {target_path} không tồn tại.")
```

### Kiểm tra cấu trúc cây và số lượng ảnh chi tiết trong thư mục `crawled`

```python
import os

target_path_crawled = os.path.join(data_path, 'crawled')
image_extensions = ('.jpg', '.jpeg', '.png', '.bmp', '.webp')

if os.path.exists(target_path_crawled):
    print(f"Cấu trúc thư mục và số lượng ảnh trong: {target_path_crawled}\n")
    count = 0
    for root, dirs, files in os.walk(target_path_crawled):
        # Đếm số ảnh trong thư mục hiện tại
        images_in_folder = [f for f in files if f.lower().endswith(image_extensions)]
        count = len(images_in_folder)
        count +=1
        # Tính độ sâu để thụt lề (visualize tree)
        level = root.replace(target_path_crawled, '').count(os.sep)
        indent = ' ' * 4 * level

        # Tên thư mục hiện tại
        folder_name = os.path.basename(root)

        if count > 0:
            print(f"{indent}├── {folder_name}/ ({count} ảnh)")
        else:
            # Chỉ in tên thư mục nếu nó có thư mục con hoặc là thư mục gốc
            if dirs or root == target_path_crawled:
                print(f"{indent}└── {folder_name}/")
    print(f"Có tất cả: {count} folder")
else:
    print(f"Thư mục {target_path_crawled} không tồn tại.")
```

Để hợp nhất dữ liệu cho việc gắn nhãn trong Label Studio, cách tốt nhất sẽ phụ thuộc vào định dạng và cấu trúc dữ liệu cụ thể của bạn. Dưới đây là một số cách tiếp cận chung:

*   **Nếu dữ liệu là hình ảnh/âm thanh:** Bạn có thể tạo một thư mục duy nhất và sao chép tất cả các tệp từ các thư mục con vào đó. Label Studio có thể tải lên các thư mục trực tiếp.

*   **Nếu dữ liệu là văn bản (ví dụ: JSON, CSV):** Bạn có thể đọc từng tệp vào DataFrame của Pandas và sau đó hợp nhất chúng lại với nhau (ví dụ: `pd.concat`). Sau đó, bạn có thể xuất dữ liệu đã hợp nhất thành một tệp JSON hoặc CSV mà Label Studio có thể nhập.

*   **Sử dụng API của Label Studio:** Để tích hợp nâng cao hơn, bạn có thể sử dụng API của Label Studio để nhập dữ liệu theo chương trình sau khi bạn đã chuẩn bị nó.

Ví dụ, nếu bạn có nhiều thư mục con chứa các tệp hình ảnh, bạn có thể sao chép chúng vào một thư mục 'merged_data' như sau (hãy đảm bảo bạn đã chạy ô trên và xác nhận các thư mục tồn tại):

```python
import os
import shutil
from pathlib import Path
from collections import defaultdict
from tqdm.auto import tqdm

# Thiết lập đường dẫn
DATA_PATH = Path('/content/drive/MyDrive/VietFood-Project/data/')
CRAWLED_DIR = DATA_PATH / 'crawled' / 'dataset' / 'raw_images'
MERGED_DIR = DATA_PATH / 'merged'
IMAGE_EXTENSIONS = ('.jpg', '.jpeg', '.png', '.bmp', '.webp')

def make_output_filename(class_name, running_id):
    """Tạo tên file theo format: class_name_000001.jpg"""
    return f"{class_name}_{running_id:06d}.jpg"

if not CRAWLED_DIR.exists():
    print(f"Không tìm thấy thư mục crawled tại: {CRAWLED_DIR}")
else:
    # 1. Khởi tạo counters dựa trên số lượng ảnh hiện có trong MERGED_DIR
    # Điều này giúp chúng ta đánh số tiếp tục mà không bị trùng
    merged_counters = defaultdict(int)
    if MERGED_DIR.exists():
        for cls_folder in MERGED_DIR.iterdir():
            if cls_folder.is_dir():
                existing_files = [f for f in cls_folder.iterdir() if f.suffix.lower() in IMAGE_EXTENSIONS]
                merged_counters[cls_folder.name] = len(existing_files)

    print("Bắt đầu gộp ảnh từ crawled vào merged...")

    # 2. Lặp qua các thư mục class trong crawled
    crawled_classes = [d for d in CRAWLED_DIR.iterdir() if d.is_dir()]

    copy_success = 0
    errors = 0

    for cls_path in tqdm(crawled_classes, desc="Đang xử lý các lớp"):
        class_name = cls_path.name
        # Tạo thư mục đích nếu chưa có
        target_cls_dir = MERGED_DIR / class_name
        target_cls_dir.mkdir(parents=True, exist_ok=True)

        # Lấy danh sách ảnh trong class này của crawled
        images = [f for f in cls_path.iterdir() if f.suffix.lower() in IMAGE_EXTENSIONS]

        for img_path in images:
            merged_counters[class_name] += 1
            new_name = make_output_filename(class_name, merged_counters[class_name])
            dst_path = target_cls_dir / new_name

            try:
                shutil.copy2(img_path, dst_path)
                copy_success += 1
            except Exception as e:
                errors += 1
                # print(f"Lỗi khi copy {img_path.name}: {e}")

    print(f"\nHoàn tất!")
    print(f"- Tổng số ảnh đã copy thành công: {copy_success}")
    print(f"- Số lỗi xảy ra: {errors}")
    print(f"- Dữ liệu hiện tại nằm tại: {MERGED_DIR}")
```
